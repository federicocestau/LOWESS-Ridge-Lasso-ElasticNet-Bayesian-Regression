```python
# FEATURE SELECTION & DATA SPLIT

X_var = df[['area', 'bedrooms', 'bathrooms', 'stories',
       'parking']].values
y_var = df['price'].values
```

```python
X_var
```

```python
#random state to receive the same values. 
X_train, X_test, y_train, y_test = train_test_split(X_var, y_var, test_size = 0.2, random_state = 0)
```

```python
import statsmodels.api as sm
```

```python
#add constant to predictor variables
x = sm.add_constant(X_train)
```

```python
#fit linear regression model
model = sm.OLS(y_train, X_train).fit()
```

```python
#view model summary
print(model.summary())
```

```python
ypred = model.predict(X_test)
print(ypred)
```

```python
# 1. LR

ols = LinearRegression()
ols.fit(X_train, y_train)
y_LR = ols.predict(X_test)
```

```python
# 2. Ridge
ridge = Ridge(alpha = 0.5)
ridge.fit(X_train, y_train)
ridge_y = ridge.predict(X_test)
```

```python
# 3. Lasso
lasso = Lasso(alpha = 0.01)
lasso.fit(X_train, y_train)
lasso_y = lasso.predict(X_test)
```

```python
# 4. Bayesian
bayesian = BayesianRidge()
bayesian.fit(X_train, y_train)
bayesian_y = bayesian.predict(X_test)
```

```python
# 5. ElasticNet
en = ElasticNet(alpha = 0.01)
en.fit(X_train, y_train)
en_y = en.predict(X_test)
```

```python
print(cl('EXPLAINED VARIANCE SCORE:', attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('Explained Variance Score of OLS model is {}'.format(evs(y_test, ypred )), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('Explained Variance Score of  Linear Regression model is {}'.format(evs(y_test, y_LR)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('Explained Variance Score of Ridge model is {}'.format(evs(y_test, ridge_y)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('Explained Variance Score of Lasso model is {}'.format(evs(y_test, lasso_y)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('Explained Variance Score of Bayesian model is {}'.format(evs(y_test, bayesian_y)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('Explained Variance Score of ElasticNet is {}'.format(evs(y_test, en_y)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
```

the best one with this metric is the OLS model. Probably is better to improve the scores adding more features and convert them to numerical.
Onehotenconder. 

```python
# 2. R-squared
print(cl('R-SQUARED:', attrs = ['bold']))
print(cl('R-Squared of OLS model is {}'.format(r2(y_test, ypred)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('R-Squared of Linear Regression model is {}'.format(r2(y_test,  y_LR)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('R-Squared of Ridge model is {}'.format(r2(y_test, ridge_y)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('R-Squared of Lasso model is {}'.format(r2(y_test, lasso_y)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('R-Squared of Bayesian model is {}'.format(r2(y_test, bayesian_y)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
print(cl('R-Squared of ElasticNet is {}'.format(r2(y_test, en_y)), attrs = ['bold']))
print('-------------------------------------------------------------------------------')
```

Similar score to the Explained Variance Score the best score is the OLS model

# Scores are not good for all regression models, but it is a good illustration of how to use all the regression variables to create different models using the same dataset.
